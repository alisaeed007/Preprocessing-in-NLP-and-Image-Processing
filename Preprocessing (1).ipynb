```python
# Start Word Tokenizer
import nltk
from nltk.tokenize import word_tokenize
#nltk.download('punkt')
text = "I like to study in UCP"
print(word_tokenize(text))
```

```python
# Start Sentence Tokenizer
import nltk
from nltk.tokenize import sent_tokenize
#nltk.download('punkt')
text = "I like to study in UCP. Do you like it"
print(sent_tokenize(text))
```

```python
# Start lower case
def punc(text):
  return text.lower()

x="Hello World"
t=punc(x)
print (t)
```

```python
# Start Remove Punctuaution
import string
def punc(text):
  r=str.maketrans('','',string.punctuation)
  return text.translate(r)

x="Hello! World, Thanks."
t=punc(x)
print (t)
```

```python
# Strat List English Stop words
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords

print(stopwords.words('english'))
```

```python
# Strat Removal of Stopwords
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

example_sent="This is my first sentence I like to study in UCP"
stop_words=set(stopwords.words('english'))

word_tokens=word_tokenize(example_sent)

filtered_sent=[]
for w in word_tokens:
  if w not in stop_words:
    filtered_sent.append(w)

print(filtered_sent)
```

```python
# Start Stemmer using PorterStemmer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
  
ps = PorterStemmer()
  
print(ps.stem("corpora"))
print(ps.stem("rocks"))

```

```python
# import these modules
#nltk.download('wordnet')
#nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

x = WordNetLemmatizer()

print(x.lemmatize("corpora"))
print(x.lemmatize("rocks"))
```
